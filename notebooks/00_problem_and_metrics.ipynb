# Voice Command Assistant (DAT255)

### Project Overview

This project aims to build a compact, low-latency deep learning model that recognizes a small set of spoken voice commands from short audio clips (≈1 second).

The system will run locally and respond immediately to microphone input.
If time permits, the model will be used to control a simple Snake game using voice commands.

### Problem Understanding (Big Picture)

Objective
- Input: Short audio clip (~1 second)
- Output: Command label (e.g., up, down, left, right, unknown, silence)
- Use case: First-stage voice assistant (wake/command recognition before cloud processing)

### Scope and Delimitations

In Scope
- Train at least one deep learning model from randomly initialized weights
- Compare different preprocessing strategies and architectures
- Perform objective evaluation on unseen test data
- Demonstrate real-time microphone inference (Gradio/Streamlit demo)

Out of scope
- Full speech-to-text (ASR)
- Large-scale streaming production system
- Complex language modeling

### Dataset

We wil use:
Google Speech Commands Dataset
https://huggingface.co/datasets/google/speech_commands

Selected Label Set
To support game control (Snake), we define:

Primary commands
- up
- down
- left
- right

Additional classes
- unknown (all other words)
- silence (background noise / no speech)

This ensures the model can handle out-of-class audio robustly.

### Success Criteria (Measurable Goals)

Primary Performance Goals
- High classification performance on unseen test data
- Good performance across all classes (balanced metric)

Secondary Engineering Goals
- Low inference latency (CPU)
- Small model size
- Robustness to speaker variation

Target Metrics (initial goals)
- Macro F1-score ≥ 0.85
- Inference latency < 50 ms per sample (CPU)
- Model size < 10 MB (optional target)

These targets may be refined after baseline experiments.

### Evaluation Metrics
We will evaluate using:
- Accuracy
- Macro F1-score (important due to class imbalance)
- Confusion matrix
- Inference latency (measured empirically)

Macro F1 is prioritized to ensure fair performance across all classes.

### Baselines

Baseline 0 – Naive
Always predict unknown.

Purpose:
- Sanity check
- Ensure learned models outperform trivial solutions

Baseline 1 – Simple CNN on Spectrogram
- 2–3 convolutional blocks
- Mel-spectrogram input

Baseline 2 (Optional)
- Conv1D model on raw waveform
- Compare raw vs spectrogram representation

### Experimental Plan
Preprocessing Comparison
- Mel-spectrogram (standard)
- (Optional) MFCC or raw waveform

Model Comparison
- ScratchCNN-Spectrogram
- Lightweight CNN variant
- Optional Conv1D raw model

Regularization & Optimization
- Dropout
- Data augmentation (time shift, background noise)
- Early stopping

### Reproducibility

We will ensure reproducibility by:
- Fixing random seeds
- Logging configurations
- Saving trained models
- Saving evaluation metrics and plots in /results

### Risks and Challenges
- Class imbalance
- Domain shift (dataset vs live microphone input)
- Data leakage during split
- Latency constraints
